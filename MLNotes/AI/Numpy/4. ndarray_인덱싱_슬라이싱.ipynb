#### 인덱싱
- 파이썬 리스트와 동일한 개념으로 사용
- ,를 사용하여 각 차원의 인덱스에 접근 가능

* 1차원 벡터 인덱싱

```python
x = np.arange(10)
print(x)
```

```python
x[0]        # 첫번째 원소를 가져옴
```

```python
x[-1]       # 마지막 원소를 가져옴
```

```python
x[3]
```

```python
x[3] = 100
print(x)
```

* 2차원 행렬 인덱싱

```python
x = np.arange(10).reshape((2, 5))
print(x)
```

```python
x[0]
```

```python
x[1]
```

```python
x[0, 2]
```

```python
x[1, 3]
```

```python
x[1, -1]
```

* 3차원 텐서 인덱싱

```python
x = np.arange(36).reshape(3, 4, 3)
print(x)
```

```python
x[0]                  # 2차원 배열을 가져옴
x[1, 2]               # 크기가 3인 벡터를 가져옴 
x[0, 2, 2]            # 스칼라 값을 가져옴
```

#### 슬라이싱
- 리스트, 문자열 slicing과 동일한 개념으로 사용
- ,를 사용하여 각 차원별로 슬라이싱 가능

* 1차원 벡터 슬라이싱

```python
x = np.arange(10)
print(x)
```

```python
x[1:7]
```

```python
x[1:]
```

```python
x[:]
```

* 2차원 행렬 슬라이싱

```python
x = np.arange(10).reshape(2,5)
print(x)
```

```python
x[:, 1:4]
```

```python
x[:, :2]
```

```python
x[0, 1:4]         # 인덱싱은 차원이 한차원 줄어듦(=> 1차원)
```

```python
x[:1, 1:4]        # 슬라이싱은 차원이 줄지 않음
```

* 3차원 텐서 슬라이싱

```python
x = np.arange(54).reshape(2, 9, 3)
print(x)
```

```python
x[:1, :2, :]          # 슬라이싱으로 가져옴(차원은 유지됨)
```

```python
x[0, :2]
x[0, :2, :]           # 인덱싱으로 가져올 때는 차원이 한차원 줄어듦 
```
