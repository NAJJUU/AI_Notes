#### Matrix(행렬) 표기법

- Matrix
$$\begin{pmatrix} 1 & 1 \\ 2 & 4 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}=\begin{pmatrix} 25 \\ 64 \end{pmatrix}$$

- Vector
  - 행(row vector) 또는 열(column vector)의 수가 1인 경우

- 역행렬(Inverse)
  - 아래를 만족할 때, B는 A의 역행렬이고 A는 B의 역행렬임
  - AB = BA = E(단위 행렬) 

#### numpy.linalg.inv
- 역행렬을 구할 때 사용하는 함수
- 모든 차원의 값이 같아야함(정방향이 되어야함)

```python
x = np.random.rand(3, 3)
print(x)

y = np.linalg.inv(x)
print(y)
```

```python
x @ np.linalg.inv(x)        # 행렬과 역행렬을 곱함 = 단위행렬(항등행렬) 
```

```python
np.matmul(x, np.linalg.inv(x))    # 행렬과 역행렬을 곱함 
```

```python
x = np.random.rand(3, 3, 3)     # 정사각행렬만 가능함(차원만 늘림)
print(x)

np.linalg.inv(x)
```

```python
np.matmul(x, np.linalg.inv(x))
```

```python
x @ np.linalg.inv(x)
```

#### numpy.linalg.solve
- Ax = B 형태의 선형대수식 솔루션을 제공하는 함수
- 예) 
  - x와 y의 합: 25
  - 2x와 4y의 합: 64
  - $$\begin{pmatrix} 1 & 1 \\ 2 & 4 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}=\begin{pmatrix} 25 \\ 64 \end{pmatrix}$$


```python
A = np.array([[1, 1], [2, 4]])
B = np.array([25, 64])

x = np.linalg.solve(A, B)
print(x)
```

```python
np.allclose(A@x, B)       # 두 수를 비교함
```
