#### 목표
- random 서브모듈 함수를 통해 ndarray 생성하기

```python
import numpy as np
```

#### rand 함수
- 0, 1 사이의 분포로 랜덤한 ndarray 생성

```python
np.random.rand(2, 3)    # 2행 3열 ndarray 생성
```

```python
np.random.rand(2)       # 1차원 벡터 ndarray 생성
```

```python
np.random.rand(10)      # 크기가 10인 1차원 벡터 ndarray 생성
```

```python
# 5행 3열인 행렬이 4개 만들어지게 ndarray 생성하기

# 내가 쓴 코드
# x = np.random.rand(60)
# x.reshape(4, 5, 3)

np.random.rand(4, 5, 3)
```

#### randn 함수
- n: normal distribution(정규분포)
- 정규분포로 샘플링된 랜덤 ndarray 생성

```python
np.random.randn(3, 4)
```

```python
import matplotlib.pyplot as plt
```

```python
y = np.random.randn(3, 4)
print(y)
plt.plot(y)
```

```python
y = np.random.randn(5)
print(y)
plt.plot(y)
```

```python
y = np.random.randn(10000)
plt.hist(y)
```

#### randint 함수
- 특정 정수 사이에서 랜덤하게 샘플링

```python
np.random.randint(1, 46, (5, 6))
```

```python
np.random.randint(1, 46, (3, 5, 6))
```

```python
# 크기가 5인 1차원 벡터 생성
# 내가 쓴 코드
np.random.randint(1, 101, 5)

np.random.randint(1, 100, size=(5,))
```

#### seed 함수
- 랜덤한 값을 동일하게 다시 생성하고자 할 때 사용

```python
np.random.seed(100)       # 씨앗으로 쓰이는 값이므로 임의의 숫자로 정하면 됨
np.random.randn(3, 4)
```

#### choice 함수
- 주어진 1차원 ndarray로부터 랜덤으로 샘플링
- 정수가 주어진 경우, np.array(해당숫자)로 간주

```python
np.random.choice(46, (2, 3))
```

```python
x = np.array([1, 2, 3, 8.55, 2.6, 8])
np.random.choice(x, (2, 3))                     # x에서 중복을 허용하여 choice
np.random.choice(x, (2, 3), replace=False)      # replace를 False로 정의하면 중복을 허용하지 않고 choice
```

#### 확률분포에 따른 ndarray
- uniform
- normal

```python
np.random.uniform(1.0, 3.0, (4, 5))         # uniform 분포
```

```python
np.random.normal((3, 4))                     # 정규분포
np.random.randn(3, 4)
```
