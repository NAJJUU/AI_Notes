```python
import numpy as np
import matplotlib.pyplot as plt
```

#### np.array 함수로 생성하기

```python
x = np.array([1, 2, 3, 4])
print(x)

y = np.array([[2, 3, 4], [1, 2, 5]])      # 2행 3열의 행렬 생성 
print(y)
print(type(y))
```

#### np.arrange 함수로 생성하기


```python
np.arange(10)         # ndarray 생성
```

```python
np.arange(1, 11)     # python의 range()와 비슷한 기능임
```

```python
np.arange(1, 11, 2)
```

```python
# 5에서 100까지 간격 5인 ndarray 생성
np.arange(5, 101, 5)
```

#### np.ones, np.zeros로 ndarray 생성하기

```python
np.ones((4, 5))
```

```python
np.ones((2, 3, 4))      # 3행 4열 행렬이 2개가 생성됨
```

```python
np.zeros((2, 3, 4, 5))
```

#### np.empty, np.full로 ndarray 생성하기

```python
np.empty((3, 4))        # 특정값으로 초기화되지 않음. 임의의 값으로 채워짐
```

```python
np.full((3, 4, 2), 8)
```

#### np.eye로 ndarray 생성하기
- 단위 행렬 생성
  - 단위 행렬은 행과 열의 수가 같음
  - 단위 행렬을 대각선은 1, 나머지는 0인 행렬임

```python
np.eye(3)         # 3행 3열 행렬을 만들면서 대각선은 1로 생성함
```

```python
np.eye(5)
```

#### np.linspace로 ndarray 생성하기

```python
np.linspace(3, 7, 3)      # 3부터 7사이 3개의 수 생성하는데 그 간격이 동일하게 만듬
```

```python
np.linspace(1, 10, 4)
```

```python
np.linspace(1, 10, 5)
```

#### reshape 함수 사용
  - ndarray의 형태, 차원을 바꾸기 위해 사용

```python
x = np.arange(1, 16)
print(x)

x.shape     # 튜플로 표시됨. 1차원이라서 1개로만 표시됨 

x.reshape(3, 5)     # 1차원이 3행 5열의 행렬로 바뀜
x.reshape(5, 3)

x.reshape(5, 1, 3)  # 3차원으로 바뀜 

# x.reshape(5, 4)   # 데이터가 최소 20개가 필요함. 그러나 15개의 데이터만 존재하기 때문에 에러 발생함
```
