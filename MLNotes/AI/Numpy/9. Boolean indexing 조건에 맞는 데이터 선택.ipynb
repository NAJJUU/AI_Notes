#### Boolean indexing
- ndarray 인덱싱 시, bool 리스트를 전달하여 True인 경우만 필터링
- 예) 짝수인 경우만 찾아보기

```python
import numpy as np
```

```python
x = np.random.randint(1, 100, size=10)
print(x)
```

```python
even_mask = x % 2 == 0
print(even_mask)
```

* bool 리스트를 인덱스로 전달

```python
x[even_mask]
```

```python
x[x%2==0]
```

```python
x[x>50]
```

#### 다중조건 사용하기
- 파이썬의 논리 연산자 and, or, not 키워드 사용 불가
- &, | 사용

```python
x % 2 == 0
x > 80

x[(x % 2 == 0) & (x > 50)]
```

```python
x < 30
x > 50

x[(x < 30) | (x > 50)]
```

예) 2023년 7월 서울 평균 기온 데이터 
  - 평균기온이 25도를 넘는 날의 수는 몇일인가?
  - 평균기온이 25도를 넘는 날의 평균 기온은 몇도인가?

```python
tmp = np.array(
    [23.9, 24.4, 26.0, 28.1, 27.6, 29.7, 26.1, 25.0, 24.6, 26.9, 27.1, 28.3,
     24.9, 26.3, 23.8, 29.3, 27.6, 21.0, 26.4, 28.0, 26.4, 24.5, 23.9, 23.9, 
     26.3, 25.9, 28.4, 27.5, 25.8, 24.9, 26.3]
)

len(tmp)
```

```python
tmp[tmp>25.0]
```

```python
len(tmp[tmp>25.0])
```

```python
tmp[tmp>25.0].mean()
```

```python
len(tmp[tmp>25.0])
```

```python
np.sum(tmp>25.0)            # tmp>25.0 조건에 맞는 개수를 반환
```

```python
np.sum(tmp[tmp>25.0])       # tmp에서 tmp>25.0 인덱싱의 값들 합을 반환 
```
