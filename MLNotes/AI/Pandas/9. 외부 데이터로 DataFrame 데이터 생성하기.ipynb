#### csv 데이터로부터 DataFrame 생성하기
- 데이터 분석을 위해서 외부파일로부터 dataframe을 생성하는 것이 가장 일반적임
- csv(comma seperated values) 파일로부터 생성
- pandas.read_csv() 함수 사용

```python
import pandas as pd
```

```python
#data : https://www.kaggle.com/datasets/hesh97/titanicdataset-traincsv
train_data = pd.read_csv('./sample_data/train.csv')     # 런타임 이후에는 삭제됨 
train_data.head()
```

#### read_csv() 함수 파라미터
- sep: 각 데이터 값을 구별하기 위한 구분자(seperator) 설정
- header: header를 무시할 경우, None 설정
- index_col: index로 사용할 column 설정
- usecols: 실제로 dataframe에 로딩할 columns만 설정

```python
train_data = pd.read_csv('./sample_data/train.csv', sep='#')
train_data      # sep로 '#'이 없으므로 전체가 하나로 데이터가 만들어짐 
```

```python
train_data = pd.read_csv('./sample_data/train.csv', sep=',')
train_data
```

```python
train_data.columns
```

```python
train_data = pd.read_csv('./sample_data/train.csv', header=None)      # 헤더를 값으로 인식하게 됨 
train_data
```

```python
train_data.columns
```

```python
# 인덱스는 컬럼이기 때문에 usecols를 사용할 시 포함시켜주어야함
train_data = pd.read_csv('./sample_data/train.csv', index_col='PassengerId', usecols=['PassengerId', 'Survived', 'Name', 'Sex', 'Age']) 
train_data            
# index_col: 어떤 컬럼을 인덱스로 쓰고자하는 경우 사용
# use_cols: 필요한 컬럼만 명시할 수 있음
```
