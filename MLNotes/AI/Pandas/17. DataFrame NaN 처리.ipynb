```python
#data : https://www.kaggle.com/datasets/hesh97/titanicdataset-traincsv
train_data = pd.read_csv('./sample_data/train.csv')     # 런타임 이후에는 삭제됨 
train_data.head()
```

#### NaN 처리 방법
- 데이터에서 삭제
  - dropna() 함수
- 다른 값으로 치환
  - fillna() 함수



- NaN 데이터 삭제하기

```python
train_data.tail()
```

```python
train_data.dropna()   # 기본적으로 row기반으로 동작함
# 어떤 데이터에 NaN값이 한개라도 있으면 그 row는 지워버림
```

```python
train_data.dropna(subset=['Age'])     # 아래 인덱스 888의 Age가 NaN이어서 해당 row는 삭제됨
```

```python
train_data.dropna(subset=['Age', 'Cabin'])    # row(행) 레벨로 지움
```

```python
train_data.dropna(axis=1)     # 컬럼이 줄어듦(컬럼에 NaN 값이 들어있으면 컬럼 자체를 삭제함)
train_data.corr()
```

* NaN값 대체하기
  - 평균으로 대체하기
  - Survived 경우 생존자/사망자 별 평균으로 대체하기

```python
train_data['Age'].fillna(train_data['Age'].mean())    # fillna()으로 'Age' 평균값으로 NaN값을 대체하기
```

```python
train_data[train_data['Survived']==1]
```

```python
# 생존자 평균 나이
mean1 = train_data[train_data['Survived']==1]['Age'].mean()
mean1

# 사망자 평균 나이
mean0 = train_data[train_data['Survived']==0]['Age'].mean()
mean0

print(mean1, mean0)
```

```python
# 생존자 나이 NaN인거 생존자의 평균나이로 대체하기
train_data[train_data['Survived']==1]['Age'].fillna(mean1)
```

```python
# 사망자 나이 NaN인거 사망자의 평균나이로 대체하기
train_data[train_data['Survived']==0]['Age'].fillna(mean0)      # 원본에는 영향을 주지않음
```

```python
train_data
```

```python
train_data.loc[train_data['Survived']==1, 'Age'] = train_data[train_data['Survived']==1]['Age'].fillna(mean1)   # 원본에 반영함
train_data.loc[train_data['Survived']==0, 'Age'] = train_data[train_data['Survived']==1]['Age'].fillna(mean0)   # 원본에 반영함
```
