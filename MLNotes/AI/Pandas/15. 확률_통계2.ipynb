```python
import numpy as np
import pandas as pd
```

```python
en_scores = [42, 69, 56, 41, 57, 48, 65, 49, 65, 58]
ma_scores = [65, 80, 63, 63, 76, 60, 81, 66, 82, 78]
```

```python
scores_df = pd.DataFrame(
    {'english':en_scores, 'math':ma_scores},
    index=pd.Index(
        ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
        name='student'
    )
)

scores_df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)

# 산점도 그리기
# 첫번째 인수가 x축, 두번째 인수가 y축
ax.scatter(scores_df['english'], scores_df['math'])
ax.set_xlabel('english')
ax.set_ylabel('math')

plt.show()
```

- Axes : 보통 plot으로 생각하는 하나의 그래프. 각각의 Axes는 개별적으로 제목 및 x/y 레이블을 가질 수 있다.
- Axis : 번역하면 '축'인데, 그려지는 축을 말하는 것이 아니라 정확히는 x, y 의 제한 범위를 말한다.
