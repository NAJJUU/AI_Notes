```python
import numpy as np
import pandas as pd
```

#### Boolean Selection
  * Series가 []와 함께 사용되면 True값에 해당하는 값만 새로 반환되는 Series 객체에 포함됨
  * 다중조건의 경우, &(and), |(or)를 사용하여 연결 가능

```python
s = pd.Series(np.arange(10), np.arange(10)+1)         # 값, 인덱스
s
```

```python
s > 5         # value(값)를 비교
```

```python
s[s>5]        # True인 것만 필터링함
```

```python
s % 2 == 0      # 짝수인 것만 가져옴
```

```python
s[s % 2 == 0]   # 인덱스의 값을 기준으로 비교함 
```

```python
# 인덱스를 기준으로 실행
s
```

```python
s.index > 5       # 인덱스를 기준으로 비교함 
```

```python
s[s.index > 5]
```

```python
(s > 5) & (s < 8)
```

```python
s[(s > 5) & (s < 8)]
```

```python
(s >= 7).sum()      # 7보다 큰것의 개수
```

```python
(s[s >= 7]).sum()   # s[7] + s[8] + s[9]
```
