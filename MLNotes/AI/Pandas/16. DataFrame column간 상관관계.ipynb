```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#data : https://www.kaggle.com/datasets/hesh97/titanicdataset-traincsv
train_data = pd.read_csv('./sample_data/train.csv')     # 런타임 이후에는 삭제됨 
train_data.head()
```

#### 변수(column) 사이의 상관계수(correlation)
- corr()함수를 통해 상관계수 연산(-1, 1 사이의 결과)
- 연속성(숫자형) 데이터에 대해서만 연산
- 인과관계를 의미하진 않음

```python
train_data.corr()

# 1에 가까울수록 두개 컬럼이 비슷하게 증가하는 것임(두개의 패턴이 비슷함)
# -1에 가까울수록 하나 증가하면 하나는 감소하는 것임
# 0에 가까우면 두 컬럼 사이는 관계가 없는 것임
# 대각선 값은 항상1임. 이는 같은 변수(컬럼)에 대해서 연산을 하기 때문임
# 이 대각선 기준으로 좌우 대칭적인 모습을 보임
```

```python
plt.matshow(train_data.corr())      # 색깔이 밝을수록 관계가 더 깊음
```
