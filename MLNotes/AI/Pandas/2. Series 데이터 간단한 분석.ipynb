```python
import numpy as np
import pandas as pd
```

#### property, function
  - size : 개수 반환
  - shape : 튜플형태로 shape 반환
  - unique() : 유일한 값만 ndarray로 반환
  - count() : NaN을 제외한 개수를 반환
  - mean() : NaN을 제외한 평균
  - value_counts() : NaN을 제외한 각 값들의 빈도를 반환

```python
s = pd.Series([1, 1, 2, 1, 2, 2, 2, 1, 1, 3, 3, 4, 5, 5, 7, np.NaN])
s
```

```python
len(s)       # Series 길이
```

```python
s.size
```

```python
s.shape      # 튜플 형태로 값을 보여줌  -- Series는 항상 1차원 
```

```python
s.unique()
```

```python
s.count()
```

```python
a = np.array([2, 2, 2, 2, np.NaN])
a.mean()        # numpy는 NaN이 있으면 평균이 구해지지 않음
```

```python
b = pd.Series(a)
b.mean()       # pandas는 NaN이 있으면  NaN을 무시하고 평균을 구해줌
```

```python
s.mean()
```

```python
s.value_counts()      # 빈도 
```

- index를 활용한 멀티플한 값에 접근가능

```python
s[5]
```

```python
s[[5, 7, 8]]        # 해당 인덱스에 있는 값을 가져옴([[]]) => 꺽쇄를 두번 사용함
```

```python
s[[5, 7, 8, 10]]
```

```python
s[[5, 7, 8, 10]].value_counts()
```
