```python
import numpy as np
import pandas as pd
```

#### index 기준으로 연산

```python
s1 = pd.Series([1, 2, 3, 4], ['a', 'b', 'c', 'd'])
s2 = pd.Series([6, 3, 2, 1], ['d', 'c', 'b', 'a']) 
s1
```

```python
s2
```

```python
s1 + s2       # 같은 인덱스끼리 연산이 이루어짐 
```

#### 산술연산
* Series의 경우에도 스칼라와의 연산은 각 원소별로 스칼라와의 연산이 적용
* Series와의 연산은 각 인덱스에 맞는 값끼리 연산이 적용
  - 이때 인덱스의 쌍(pair)이  맞지 않으면 결과는 NaN

```python
s1 ** 2
```

```python
s1 ** s2        # d의 경우 4의 6승
```

```python
4 ** 6
```

#### index pair가 맞지 않는 경우의 연산
