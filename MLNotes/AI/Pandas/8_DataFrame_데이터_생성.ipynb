#### DataFrame 생성하기
- 다른 데이터 소스(database, 외부파일)을 통해 dataframe 생성
- dummy 데이터를 생성하는 방법

```python
import pandas as pd
```

#### dictionary로부터 생성하기
- dict의 key -> column

```python
data = {'a' : 100, 'b' : 200, 'c' : 300}
pd.DataFrame(data, index=['x', 'y', 'z'])
```

```python
data = {'a' : [1, 2, 3], 'b' : [4, 5, 6], 'c' : [10, 11, 12]}
pd.DataFrame(data, index=[0, 1, 2])
```

#### Series로부터 생성하기
- 각 Series의 인덱스 -> Column

```python
a = pd.Series([100, 200, 300], ['a', 'b', 'c'])
b = pd.Series([10, 20, 30], ['a', 'b', 'c'])
c = pd.Series([1, 2, 3], ['a', 'b', 'c'])

pd.DataFrame([a, b, c], index=[100, 101, 102])
```

```python
a = pd.Series([100, 200, 300], ['a', 'b', 'c'])
b = pd.Series([10, 20, 30], ['a', 'b', 'd'])
c = pd.Series([1, 2, 3], ['a', 'b', 'k'])

pd.DataFrame([a, b, c], index=[100, 101, 102])
```
