```python
import numpy as np
import pandas as pd 
```

#### Series 값 업데이트
- 추가 및 변경: 인덱스를 이용
- 삭제: drop() 함수 이용

```python
s = pd.Series(np.arange(100, 105), ['a', 'b', 'c', 'd', 'e'])
s
```

```python
s['a'] = 200
s
```

```python
s['k'] = 300        # 해당 인덱스는 추가됨
s
```

```python
s.drop('k')     # 'k'를 삭제, 삭제되는 객체가 따로 생성됨(=원본에 영향을 주지 않음), 복사본이 생성되고 원본은 그대로 있게됨 
```

```python
s             # 원본 s에는 아무런 영향을 주지 않음 
```

```python
s.drop('k', inplace=True)     # inplace=True를 실행하면 반환값이 없음, s에 바로 연산을 함(원본에 반영됨)
```

```python
s                             # k가 삭제된 것을 확인할 수 있음 
```

```python
s[['a', 'b']] = [300, 900]        # 한번에 업데이트 가능 
s
```

#### Slicing
- 리스트, ndarray와 동일하게 적용

```python
s1 = pd.Series(np.arange(100, 105))
s1
```

```python
s1[1:3]         # 인덱스 숫자기반인 경우의 슬라이싱
```

```python
s2 = pd.Series(np.arange(100, 105), ['a', 'b', 'c', 'd', 'e'])        # 인덱스가 문자열인 경우 
s2
```

```python
s2[1:3]         # 인덱스가 문자열인 경우도 역시 슬라이싱이 됨 
```

```python
s2['b':'d']     # 인덱스가 문자열인 경우의 슬라이싱(인덱스를 문자열로 함) => 마지막까지 포함
```
