#### dataframe slicing
- dataframe의 경우 기본적으로 [] 연산자가 column 선택에 사용
- 그러나, slicing은 row 레벨로 지원함

```python
train_data[7:10]
```

#### row 선택하기
- Series의 경우 []로 row 선택이 가능하나, DataFrame의 경우는 기본적으로 column을 선택하도록 설계됨
- .loc, .iloc로 row 선택가능
  - loc: 인덱스 자체를 사용
  - iloc: 0 based index로 사용
  - 이 두 property(loc, iloc)는 ,를 사용하여 column 선택도 가능

```python
train_data.head()
```

```python
train_data.info()
```

```python
train_data.describe()
```

```python
train_data.index = np.arange(100, 991)
train_data
```

```python
train_data.head()
```

```python
train_data.tail()
```

```python
train_data.loc[986]     # 인덱스가 986번 하나만 Series 형태로 가져옴
```

```python
train_data.loc[[210, 480, 990]]       # 복수의 row를 가져오려면 리스트 형태로 실행
```

```python
# train_data.loc[[210, 480, 990, 10000]]        # 존재하지 않는 index 사용시 에러발생
```

```python
train_data.iloc[0]
```

```python
train_data.iloc[[0, 100, 108, 2]]             # 리스트 형태로 실행 가능함
```

#### row, column 동시에 선택하기
- loc, iloc 속성을 이용할 때 ,를 이용해서 둘 다 명시 가능

```python
train_data.loc[[886, 100 ,101, 890], ['Survived', 'Name', 'Sex', 'Age']]      # row, column 순서대로 함
```

```python
train_data.iloc[[101, 110, 200, 102], [1, 3, 4, 5]]       # 존재하는 row와 column들만 명시해서 요청
```

```python
train_data.iloc[np.random.randint(991, size = 5), [1, 3, 4, 5]]     # [np.random.randint()]로 사용하면 오류가 나는데 np.random.randint() 자체가 리스트로 반환해서인듯
```
