```python
# 지역변수, 전역변수
# 전역변수
x=30
y=40

# 지역변수
def temp(x, y):
  print(x, y)
  result = x + y 
  return result

temp(10, 5)
print(x, y)
```

```python
# 리턴 값이 2개 이상
def mul_div(x, y):
  m = x * y
  d = x / y 
  return m, d 

mul_div(10, 5)
print(type(mul_div(5,10)))
print(mul_div(5,10))
```

```python
# 가변 파라미터
# 전달되는 파라미터의 개수가 고정적이지 않은 경우 

def temp(*args):
  return args

temp(10)
temp(10, 20)
print(type(temp(10,20,30,40,50)))
temp(10,20,30,40,50)
```

```python
# 키워드 가변 파라미터

def temp(**kwargs):
  print(kwargs)
  for k, v in kwargs.items():
    print(k, v)

temp(a=1, b=2, c=3)
```

```python
# 워밍문제1
# "GPT(Geneative Predefined Transformer)" 문자열을 10번 출력하는 print_gpt 합수를 작성하시오.

def print_gpt():
  for a in range(10):
    print("GPT(Geneative Predefined Transformer)")

print_gpt()
```

```python
# 워밍문제2
# 입력값 자연수가 홀수인지 짝수인지 판별하는 함수를 작성하시오. 
# 함수이름은 odd_even()
# 출력: 짝수 or 홀수 or 잘못된 값입니다.

def odd_even(a):
  if a%2==0:
    print("짝수")
  elif a%2==1:
    print("홀수")
  else :
    print("잘못된 값입니다.")

odd_even(7)
```
