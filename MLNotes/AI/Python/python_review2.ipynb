```python
# 클래스 사용(init 생성자 사용)
class Cake:
  def __init__(self):   # self: 자기가 가리키고 있는 주소, 즉 자기 자신 
    self.fruit = "포도"
    self.snack = "콘칩"

a = Cake()  
b = Cake()
print(a.fruit)
print(b.snack)

a.fruit = "딸기"
print(a.fruit)
```

```python
class Cake:
  def __init__(self, fruit, snack):
    self.fruit = fruit
    self.snack = snack

a = Cake("딸기", "치토스")
b = Cake("레몬", "오레오")
print(a.fruit, a.snack)
print(b.fruit, b.snack)
```

```python
class Cake:
  def __init__(self, fruit, snack, price):
    self.fruit = fruit
    self.snack = snack 
    self.price = price

  def sell(self):
    print("이 케익은 {}원에 팔았습니다.".format(self.price))

a = Cake("리치", "포카칩 어니언맛", "2500")
a.sell()
```

```python
class Cake:
  def __init__(self, fruit, snack, price):
    self.fruit = fruit
    self.snack = snack 
    self.price = price
    self.total = 0

  def sell(self):
    self.total += self.price
    print("이 케익은 {}원에 팔았습니다.".format(self.price))
  
  def gain(self):
    print("총 케익 판매액은 {}원입니다.".format(self.total))

a = Cake("리치", "포카칩 어니언맛", 2500)
a.sell()
a.sell()
a.gain()

b = Cake("포도", "쵹쵹한 쵸코칩", 5000)
b.sell()
b.sell()
b.sell()
b.gain()
```

```python
# 클래스 상속
# 상속을 하려고하는 클래스 = parent
# 상속을 받는 새로운 클래스 = childe

# 부모 클래스
class Car:
  def __init__(self, name, km):
    self.name = name
    self.km = km

  def paint(self, color):
    print("{}차 색상은 {}입니다.".format(self.name, color))

  def order(self, day):
    print("{}차의 출고기간은 {}일 입니다.".format(self.name, day))
  
# 자식 클래스
class Super(Car):
  def __init__(self, name, km):
    self.name = name
    self.km - km 

a = Car("현기", "500")
a.paint("황금색")
a.order(90)

b = Super("창세기","300")
b.paint("검은색")
b.order(100)
```
