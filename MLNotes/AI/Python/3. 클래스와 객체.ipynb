#### **self**
- 파이썬의 method는 항상 첫번째 인자로 self를 전달
- self는 현재 해당 메서드가 호출되는 객체 자신을 가리킴
- Java, C++/C#의 this에 해당
- 위치는 항상 맨 처음의 parameter이며 이름을 관례적으로 self로 사용

```python
class Person:
  def __init__(self, n, a):
    self.name=n
    self.age=a

  def sleep(self):
    print(self.name, '은 잠을 잡니다.')

p1=Person('차무식', 40)
p2=Person('마동석', 50)

p1.sleep()
p2.sleep()
```

```python
# Counter
# 멤버함수 추가 - 숫자를 하나 증가(increment)
#                 숫자를 0으로 초기화(reset)
#                 현재 숫자값 출력(print_current_value)

class Counter:
  def __init__(self, num=0):
    self.num = num

  def increment(self):
    self.num +=1

  def reset(self):
    self.num = 0

  def print_current_value(self):
    print('현재 값은: ',self.num)


a = Counter()
b = Counter(8)

a.increment()
a.print_current_value()

b.increment()
b.print_current_value()

a.reset()
a.print_current_value()
```

#### **method type**
- instance method - 객체로 호출
  - 메서드는 객체 레벨로 호출되기 때문에, 해당 메서드를 호출한 객체에만 영향을 줌
- Class method(static method)
  - 클래스 메서드의 경우 클래스 레벨로 호출됨


```python
class Math:
  @staticmethod
  def add(a, b):
    return a+b 

  @staticmethod
  def multiply(a, b):
    return a*b

# a = Math()
# b = Math()
# a.add(10, 20)
# a.multiply(10, 20)
Math.add(2, 8)
```

#### **Class Ingeritance(상속)**
  - 코드를 재사용할 수 있게 됨
  - 상속 받고자 하는 대상인 기존 클래스를 Parent, Super, Base class라고 부름
  - 상속 받는 새로운 클래스를 Child, Sub, Derived class라고 부름
  - 의미적으로 is-a관계를 갖음

```python
class Person:
  def __init__(self, n, a):
    self.name=n
    self.age=a

  def eat(self, food):
    print('{}는 {}를 먹습니다.'.format(self.name, food))

  def sleep(self, minute):
    print('{}는 {}분동안 잡니다.'.format(self.name, minute))

  def work(self, minute):
    print('{}는 {}분동안 일을 합니다.'.format(self.name, minute))

class Student(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

class Employee(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

cha = Person('차무식', 50)
cha.eat('chicken')
cha.sleep(30)
cha.work(60)

oh = Student('오승훈', 40)
oh.eat('beef')
oh.sleep(20)
oh.work(70)

lee = Employee('이상구', 30)
lee.eat('pork')
lee.sleep(10)
lee.work(90)
```

#### **method override**
  - 부모 클래스의 method를 재정의(override)
  - 하위 클래스(자식 클래스)의 인스턴스로 호출시, 재정의된 메서드가 호출됨

```python
class Person:
  def __init__(self, n, a):
    self.name=n
    self.age=a

  def eat(self, food):
    print('{}는 {}를 먹습니다.'.format(self.name, food))

  def sleep(self, minute):
    print('{}는 {}분동안 잡니다.'.format(self.name, minute))

  def work(self, minute):
    print('{}는 {}분동안 일을 합니다.'.format(self.name, minute))

class Student(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def work(self, minute):
    print('{}은 {}분동안 공부합니다.'.format(self.name, minute))

class Employee(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def work(self, minute):
    print('{}은 {}분동안 업무를 합니다.'.format(self.name, minute))

lee = Student('이순신', 40)
lee.eat('김밥')
lee.sleep(30)
lee.work(60)

cha = Employee('차무식', 50)
cha.work(20)
```

```python
class Person:
  def __init__(self, n, a):
    self.name=n
    self.age=a

  def eat(self, food):
    print('{}는 {}를 먹습니다.'.format(self.name, food))

  def sleep(self, minute):
    print('{}는 {}분동안 잡니다.'.format(self.name, minute))

  def work(self, minute):
    print('{}는 {}분동안 일을 합니다.'.format(self.name, minute))

class Student(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def work(self, minute):
    super().work(minute)
    print('{}은 {}분동안 공부합니다.'.format(self.name, minute))

class Employee(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def work(self, minute):
    super().work(minute)
    print('{}은 {}분동안 업무를 합니다.'.format(self.name, minute))

lee = Student('이순신', 40)
lee.eat('김밥')
lee.sleep(30)
lee.work(60)

cha = Employee('차무식', 50)
cha.work(20)
```

#### **special method**
  - __로 시작 __로 끝나는 특수 함수
  - 해당 메서드들을 구현하면, 커스텀 객체에 여러가지 파이썬 내장 함수나 연산자를 적용가능
  - 오버라이딩 가능한 함수 목록은 아래 링크 참조
    - https://docs.python.org/3/reference/datamodel.html
    - https://docs.python.org/3/reference/datamodel.html#obect.__add__

```python
'''
  Point
  2차원 좌표평면 각 점 (x, y)
  연산 - 두점의 덧셈, 뺄셈(1, 2) + (3, 4) = (4, 6)
         한점과 숫자의 곱셈(1, 2) * 3 = (3, 6)
'''

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def print_point(self):
    return '{}, {}'.format(self.x, self.y)
    
p1 = Point(3, 4)
p2 = Point(2, 7)

print([1, 2, 3, 4])
print(p1)
print(p2)
p1.print_point()
p2.print_point()
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __str__(self):
    return '{}, {}'.format(self.x, self.y)

  def add(self, pt):                  # p1 + p2 를 하면 오류남 
    new_x = self.x + pt.x
    new_y = self.y + pt.y
    return Point(new_x, new_y)
    
p1 = Point(3, 4)
p2 = Point(2, 7)

print([1, 2, 3, 4])
print(p1)
print(p2)

print(p1.add(p2))

print(p1+p2)
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __str__(self):
    return '{}, {}'.format(self.x, self.y)

  def __add__(self, pt):                  # p1 + p2 를 해도 오류가 안남 
    new_x = self.x + pt.x
    new_y = self.y + pt.y
    return Point(new_x, new_y)

  def __sub__(self, pt):
    new_x = self.x - pt.x
    new_y = self.y - pt.y
    return Point(new_x, new_y)

  def __mul__(self, a):
    new_x = self.x * a
    new_y = self.y * a
    return Point(new_x, new_y)

  def __getitem__(self, index):
    if index==0:
      return self.x
    elif index == 1:
      return self.y
    else:
      return -1

    
p1 = Point(3, 4)
p2 = Point(2, 7)

print([1, 2, 3, 4])
print(p1)
print(p2)

print(p1+p2)

p3 = p1 + p2
print(p3)

p4 = p1  - p2
print(p4)

p5 = p1 * 3
print(p5)

print(p1)
print(p1[0])
print(p1[1])
```
